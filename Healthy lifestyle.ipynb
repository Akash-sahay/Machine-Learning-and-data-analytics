```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:\\Users\\Akash Sahay\\Downloads\\Train_Data.csv")
```

```python
data
```

```python
data.drop(['ID1', 'ID2'], axis = 1, inplace= True)   
```

```python
data
```

```python
x_train = data.iloc[:, 0:15]
x_train['Specific ailments'].isna().value_counts()
```

```python
y_train = data['Healthy']
y_train
```

```python
x_train['Food preference'].fillna("Not Specified", inplace =True)
```

```python
x_train['Follow Diet'].fillna(method='bfill', inplace=True)
x_train['Physical activity'].fillna(method='bfill', inplace=True)
x_train['Regular sleeping hours'].fillna(method='bfill', inplace=True)
x_train['Alcohol consumption'].fillna(method='bfill', inplace=True)
x_train['Social interaction'].fillna(method='bfill', inplace=True)
x_train['Taking supplements'].fillna(method='bfill', inplace=True)
x_train['Mental health management'].fillna(method='bfill', inplace=True)
x_train['Illness count last year'].fillna(method='bfill', inplace=True)
```

```python
x_train.drop(['Any heriditary condition?'], axis = 1, inplace = True)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer 
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', OneHotEncoder(sparse = False, drop = 'first'),['Food preference']),
    ('tnf2', OrdinalEncoder(categories=[['NO','Cannot say', 'YES'],['URBAN', 'RURAL']]),['Smoker?','Living in?'])
], remainder='passthrough')
```

```python
x_train_transformed=transformer.fit_transform(x_train)
```

```python
x_train_transformed
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train_transformed, y_train)
```

```python
test_data = pd.read_csv("C:\\Users\\Akash Sahay\\Downloads\\Test_Data.csv")
```

```python
test_data
```

```python
x_test = test_data
x_test
```

```python
x_test['Follow Diet'].fillna(method='bfill', inplace=True)
x_test['Physical activity'].fillna(method='bfill', inplace=True)
x_test['Regular sleeping hours'].fillna(method='bfill', inplace=True)
x_test['Alcohol consumption'].fillna(method='bfill', inplace=True)
x_test['Social interaction'].fillna(method='bfill', inplace=True)
x_test['Taking supplements'].fillna(method='bfill', inplace=True)
x_test['Mental health management'].fillna(method='bfill', inplace=True)
x_test['Illness count last year'].fillna(method='bfill', inplace=True)
x_test['Food preference'].fillna("Not Specified", inplace =True)
```

```python
x_test_transformed=transformer.fit_transform(x_test)
```

```python
y_pred=clf.predict(x_test_transformed)
```

```python
y_pred
```

```python
df = pd.DataFrame(y_pred, columns=['predictions'])
```
