```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.metrics import confusion_matrix 
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("C:\\Users\\Akash Sahay\\OneDrive\\Desktop\\Machine_Learning_Course\\ML Day11 7\\Social_Network_Ads.csv")
df
```

```python
x= df.iloc [:,2:4].values
print (x) 
#independent variables
```

```python
y = df.iloc[:,4].values
print (y)
# dependent variable
```

```python
plt.scatter (x[y == 0,0],x[y == 0,1] ,s = 100, c='r', label = 'Cluster 1')
plt.scatter (x[y == 1,0],x[y == 1,1] ,s = 100, c='b', label = 'Cluster 2')
plt.show()
```

```python
#feature scaling of x
from sklearn.preprocessing import StandardScaler
sc_x= StandardScaler()
x = sc_x.fit_transform(x)
print (x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors =5 , metric = 'minkowski', p= 2)  # eucledian distance
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
df_x = pd.DataFrame(x_test,columns= ['Age','Salary'])
df_y = pd.DataFrame(y_test,columns = ['Y'])
df_ypred = pd.DataFrame(y_pred,columns = ['Y Pred'])
result = pd.concat ([df_x,df_y,df_ypred],axis = 1 )
print (result.head(50))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score (y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report (y_test,y_pred))
```

```python
cm = confusion_matrix (y_test,y_pred)
print (cm)	
```

```python
# lets try with k=1
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors =1 , metric = 'minkowski', p= 1)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x)
y_pred
```

```python
accuracy_score (y,y_pred)
#we see a high accuracy
```

```python
cm = confusion_matrix (y,y_pred)
print (cm)	
```
