```python
from sklearn.model_selection import train_test_split

train,test = train_test_split(df,train_size=0.6)
```

```python
train
```

```python
test
```

```python
x_train =train.iloc[:, 1:3001]
x_train
```

```python
x_test =test.iloc[:, 1:3001]
x_test
```

```python
ytrain =train['Prediction']
ytrain
```

```python
ytest = test['Prediction']
ytest
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(x_train, ytrain)
y_pred = mnb.predict(x_test)
y_pred
```

```python
from sklearn import metrics
metrics.accuracy_score(ytest, y_pred)            # thus we have an accuracy of 94.24 %
```

```python
from sklearn import svm
modl = svm.SVC()
modl.fit(x_train, ytrain)
svm_prediction = modl.predict(x_test)
svm_prediction
```

```python
from sklearn import metrics
metrics.accuracy_score(ytest, svm_prediction)  # thus we have a 79.41% accuracy in case of svm
                                            # thus we will use multinomial Naive Bayes Classifier for this particular classification problem
```
